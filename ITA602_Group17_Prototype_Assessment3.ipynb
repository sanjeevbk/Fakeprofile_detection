```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import tensorflow as tf
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
# Load data
train_df = pd.read_csv('insta_train.csv')
test_df = pd.read_csv('insta_test.csv')
```

```python
# Data preprocessing
def preprocess_data(df):
    # Handle missing values, encoding, and scaling
    return df
```

```python
train_df = preprocess_data(train_df)
test_df = preprocess_data(test_df)
```

```python
# Feature engineering and selection
def select_features(df):
    # Select relevant features
    return df
```

```python
x_train = select_features(train_df).drop(columns=['fake'])
y_train = train_df['fake']
x_test = select_features(test_df).drop(columns=['fake'])
y_test = test_df['fake']
```

```python
# Scale data
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print(y_train.shape)
print(x_train.shape)
```

```python
def create_neural_network():
    model = tf.keras.models.Sequential([
        tf.keras.layers.Dense(50, input_dim=11, activation='relu'),
        tf.keras.layers.Dense(150, activation='relu'),
        tf.keras.layers.Dense(1, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
nn_model = create_neural_network()
nn_model.fit(x_train, y_train, epochs=20, verbose=1, validation_split=0.1)
```

```python
# SVM Model
svm_model = SVC(kernel='rbf', C=1, gamma='auto')
svm_model.fit(x_train, y_train)

y_pred_nn = nn_model.predict(x_test).argmax(axis=1)  # Convert to continuous
y_pred_svm = svm_model.predict(x_test)
```

```python
print("Neural Network Classification Report:")
print(classification_report(y_test, y_pred_nn))

print("SVM Classification Report:")
print(classification_report(y_test, y_pred_svm))

plt.figure(figsize=(10, 10))
cm_svm = confusion_matrix(y_test, y_pred_svm)
sns.heatmap(cm_svm, annot=True)
plt.title("Support Vector Machine Confusion Matrix")
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Load data
instagram_df_train = pd.read_csv('insta_train.csv')
instagram_df_test = pd.read_csv('insta_test.csv')
```

```python
# Data exploration
def explore_data(df):
    print(df.info())
    print(df.describe())
    print(df.isnull().sum())
    sns.countplot(x='fake', data=df)
    plt.show()

explore_data(instagram_df_train)
explore_data(instagram_df_test)
```

```python
# Preprocess data
x_train = instagram_df_train.drop(columns=['fake'])
y_train = instagram_df_train['fake']
x_test = instagram_df_test.drop(columns=['fake'])
y_test = instagram_df_test['fake']

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
# Train random forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)
```

```python
# Evaluate models
y_pred_rf = rf_model.predict(x_test)

print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))
```

```python
# Plot confusion matrices
plt.figure(figsize=(10, 10))
cm_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm_rf, annot=True)
plt.title("Random Forest Confusion Matrix")
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
train_df = pd.read_csv('insta_train.csv')
test_df = pd.read_csv('insta_test.csv')
```

```python
explore_data(train_df)
explore_data(test_df)
```

```python
x_train = train_df.drop(columns=['fake'])
y_train = train_df['fake']
x_test = test_df.drop(columns=['fake'])
y_test = test_df['fake']
```

```python
models = [
    ('Logistic Regression', LogisticRegression())
]
```

```python
for name, model in models:
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy']) if name == 'Neural Network' else None
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    y_pred_class = np.argmax(y_pred, axis=1) if name == 'Neural Network' else y_pred
    print(f"{name} Classification Report:")
    print(classification_report(y_test, y_pred_class))
    plt.figure(figsize=(10, 10))
    cm = confusion_matrix(y_test, y_pred_class)
    sns.heatmap(cm, annot=True)
    plt.title(f"{name} Confusion Matrix")
    plt.show()
```
